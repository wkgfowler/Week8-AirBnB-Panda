```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('../files/AB_NYC_2019.csv')
air_bnb
```

```python
# How many neighborhood groups are available and which shows up the most?

air_bnb_group = pd.DataFrame.from_dict({
    'ID': air_bnb['id'],
    'Neighbourhood Group': air_bnb['neighbourhood_group']
})
air_bnb_group
air_bnb_group_count = air_bnb_group.groupby(['Neighbourhood Group']).count().sort_values(by=['ID'],ascending = False)
air_bnb_group_count
```

```python
# Are private rooms the most popular in manhattan?

air_bnb_man = air_bnb[air_bnb['neighbourhood_group'] == 'Manhattan']
air_bnb_man_room = pd.DataFrame.from_dict({
    'ID': air_bnb_man['id'],
    'Room-Type': air_bnb_man['room_type']
})

air_bnb_man_room_count = air_bnb_man_room.groupby(['Room-Type']).count().sort_values(by=['ID'],ascending = False)
air_bnb_man_room_count
```

```python
air_bnb[['host_name', 'host_id']]
```

```python
# Which hosts are the busiest and based on their reviews?

air_bnb_hosts = pd.DataFrame.from_dict({
    'Host Name': air_bnb['host_name'],
    'Host ID': air_bnb['host_id'],
    'Reviews': air_bnb['number_of_reviews']
})

#air_bnb_hosts_copy = air_bnb_hosts.drop_duplicates('Host ID').reset_index(drop=True)
#air_bnb_hosts_copy
air_bnb_hosts_copy_count = air_bnb_hosts.groupby(['Host ID', 'Host Name']).sum(['Reviews']).sort_values(by=['Reviews'], ascending=False)
air_bnb_hosts_copy_count
```

```python
#Which neighorhood group has the highest average price?
air_bnb_price = pd.DataFrame.from_dict ({
    "Neighbourhood Group": air_bnb['neighbourhood_group'],
    "Price": air_bnb['price']
})
air_bnb_price_count = air_bnb_price.groupby(['Neighbourhood Group']).mean(['Price']).sort_values(by=['Price'],ascending = False)
air_bnb_price_count
```

```python
# Which neighbor hood group has the highest total price?
air_bnb_maxprice = pd.DataFrame.from_dict ({
    "Neighbourhood Group": air_bnb['neighbourhood_group'],
    "Price": air_bnb['price']
})
air_bnb_maxprice_count = air_bnb_price.groupby(['Neighbourhood Group']).max(['Price']).sort_values(by=['Price'],ascending = False)
air_bnb_maxprice_count
```

```python
#Which top 5 hosts have the highest total price?
air_bnb_price2 = air_bnb[['id', 'host_name', 'price']]
air_bnb_price2_top5 = air_bnb_price2.sort_values(by='price',ascending = False)
air_bnb_price2_top5.head()
```

```python
# What host has the highest total of prices and where are they located?

air_bnb_highhost = air_bnb[['host_id', 'host_name', 'neighbourhood_group', 'price']]
air_bnb_highhost_final = air_bnb_highhost.groupby(['host_id', 'host_name', 'neighbourhood_group']).sum(['price']).sort_values(by=['price'], ascending=False)
air_bnb_highhost_final
```
